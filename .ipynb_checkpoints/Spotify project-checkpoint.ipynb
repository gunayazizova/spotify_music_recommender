```python
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id,market="GB")
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
import pandas as pd

def process_playlist(playlist_id):
    big_playlist_tracks = get_playlist_tracks("Spotify", playlist_id)
    
    # Extract track IDs
    ids = [track['track']['id'] for track in big_playlist_tracks if track['track'] and track['track']['id']]
    
    # Chop IDs into chunks of 100
    chunk_size = 100
    ids_chopped = [ids[i:i + chunk_size] for i in range(0, len(ids), chunk_size)]
    
    # Fetch audio features and collect them
    af_list = []
    for chunk in ids_chopped:
        af_list.extend(sp.audio_features(chunk))
    
    # Convert to DataFrame and return
    return pd.DataFrame(af_list)

# Example usage:
playlist_ids = ["3KkWikt9tTxih7TrT0ahbV", "18vUeZ9BdtMRNV6gI8RnR6", "4k9DMvez9p7UWTeV2sg5pn", "4HU33McSZ93lPnmTy0LyhK", "37i9dQZF1EQqA6klNdJvwx", "3mt2ysylQQ2cccDPyGfKID", "37i9dQZF1DWV7EzJMK2FUI"]
df_all = pd.concat([process_playlist(playlist_id) for playlist_id in playlist_ids], ignore_index=True)

# Display the combined DataFrame
df_all.to_csv('df_all.csv', index=False)
```

```python
# List of desired audio features
selected_features = [
    'danceability', 'energy', 'key', 'loudness', 'mode',
    'speechiness', 'acousticness', 'instrumentalness',
    'liveness', 'valence', 'tempo'
]

# Filter the DataFrame to include only the selected features
df = df_all[selected_features]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Scale the numeric data
scaler = StandardScaler()
scaler.fit(df)
X_scaled = scaler.transform(df)
X_scaled_df = pd.DataFrame(X_scaled, columns=df.columns)

# Train the KMeans model
kmeans = KMeans(n_clusters=7, random_state=123)
kmeans.fit(X_scaled_df)

# Predict clusters
clusters = kmeans.predict(X_scaled_df)

# Analyze cluster distribution
cluster_counts = pd.Series(clusters).value_counts().sort_index()

print("Cluster counts:")
print(cluster_counts)
```

```python
K = range(2, 21)
inertia = []

for k in K:
    print("Training a K-Means model with {} clusters! ".format(k))
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(X_scaled_df)
    inertia.append(kmeans.inertia_)

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
plt.show()
```

```python
import pandas as pd
import pickle
import random
from fuzzywuzzy import process
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from IPython.display import IFrame, display

# Load the KMeans model and scaler
with open('model_and_scaler.pickle', 'rb') as f:
    loaded_data = pickle.load(f)

loaded_kmeans = loaded_data['kmeans']

# Spotify setup
client_credentials_manager = SpotifyClientCredentials(client_id='your_client_id', client_secret='your_client_secret')
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

# Load your df_billboard and df_all DataFrames here
# df_billboard = pd.read_csv('billboard_data.csv')
# df_all = pd.read_csv('all_songs_data.csv')

# Fuzzy matching for song title
def find_best_match(song_name, song_list, threshold=90):
    """
    Find the closest match for the song name using fuzzy matching.
    """
    best_match, score = process.extractOne(song_name, song_list)
    if score >= threshold:
        return best_match
    return None

# Spotify song ID retrieval
def find_song_id(song_name):
    """
    Fetch the Spotify song ID using the Spotify API.
    """
    results = sp.search(q=song_name, type='track', limit=1)
    if results['tracks']['items']:
        song_id = results['tracks']['items'][0]['id']
        return song_id
    return None

# Function to play a song from Spotify using IFrame
def play_song(track_id):
    """
    Embeds the Spotify player for a given track using the track ID.
    """
    return IFrame(src=f"https://open.spotify.com/embed/track/{track_id}",
                  width=320,
                  height=80,
                  frameborder="0",
                  allowtransparency="true",
                  allow="encrypted-media")

# Main recommendation function
def recommend_song(user_input):
    user_input = str(user_input)  # Ensure input is treated as string

    # Step 1: Display the user's input
    display(f"User input: {user_input}")

    # Step 2: Check if the song is in df_billboard
    song_list = df_billboard['Song Title'].tolist()
    best_match = find_best_match(user_input, song_list)
    
    if best_match:
        song_info = df_billboard[df_billboard['Song Title'] == best_match]
        print(f"Found song: {song_info['Song Title'].values[0]} by {song_info['Artist'].values[0]}")
        
        # Recommend a random song from df_billboard
        random_song = df_billboard.sample(1)
        print(f"Recommended song: {random_song['Song Title'].values[0]} by {random_song['Artist'].values[0]}")
        return

    # Step 3: Check for Spotify ID and Cluster if song is not found
    song_id = find_song_id(user_input)
    
    if not song_id or song_id not in df_all['id'].values:
        print("Sorry, your song is not popular.")
        return

    # Step 4: Find the cluster of the song
    song_features = df_all[df_all['id'] == song_id][[
        'danceability', 'energy', 'key', 'loudness', 'mode',
        'speechiness', 'acousticness', 'instrumentalness',
        'liveness', 'valence', 'tempo'
    ]]

    cluster = loaded_kmeans.predict(song_features)
    df_all['cluster'] = loaded_kmeans.predict(df_all[[
        'danceability', 'energy', 'key', 'loudness', 'mode',
        'speechiness', 'acousticness', 'instrumentalness',
        'liveness', 'valence', 'tempo'
    ]])

    # Step 5: Recommend a random song from the same cluster
    same_cluster_songs = df_all[df_all['cluster'] == cluster[0]]
    recommended_song = same_cluster_songs.sample(1)

    # Get Spotify track details for recommended song
    recommended_song_id = recommended_song['id'].values[0]
    recommended_song_info = sp.track(recommended_song_id)

    print(f"Recommended song from the same cluster: {recommended_song_info['name']} by {recommended_song_info['artists'][0]['name']}")

    # Display the recommended song in the Spotify embedded player
    play_song_output = play_song(recommended_song_id)

    # Display both the input and the embedded player for the recommended song
    display(play_song_output)

# Example usage
user_input = input("Enter a song name: ")
recommend_song(user_input)
```
